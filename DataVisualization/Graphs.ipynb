```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Downloads/decsynth.csv")
kept = data[["Encodings", "Coverage thresh", "Accuracy", "F-score", "AUC"]]

test = kept.set_index(["Encodings", "Coverage thresh"])
melted = pd.melt(data, id_vars=["Encodings", "Coverage thresh"], value_vars=["Accuracy", "AUC", "F-score"])
sns.set_style("whitegrid")
ax = sns.catplot(x="Encodings", col="Coverage thresh", y="value", hue="variable", data=melted, kind="bar")
ax.set(ylim=(0, 1))
#ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
ax.set_xticklabels(rotation=30)
#plt.tight_layout()

plt.subplots_adjust(top=0.88)
ax.fig.suptitle('Synthetic XRAY results')
plt.show()
```

```python
from matplotlib import rcParams

# figure size in inches
sns.set(rc={'figure.figsize':(12.7,6.27)})

data = pd.read_csv("Downloads/sepsisconstraint.csv")
kept = data[(data["Coverage thresh"] == 15)][["Encodings", "Accuracy", "Accuracy std", "F-score", "F-score std","AUC", "AUC std"]]

kept

#test = kept.set_index("Encodings")
melted = pd.melt(kept, id_vars=["Encodings"], value_vars=["Accuracy", "AUC", "F-score"])
sns.set_style("whitegrid")
melted = melted.rename(columns = {"variable" : "metric"})

ax = sns.barplot(x="metric", y="value", hue="Encodings", data=melted)
ax.set(ylim=(0.75, 1))
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
#plt.tight_layout()
plt.legend(loc='lower right', prop={'size':12})
plt.title('Sepsis Declare results')
plt.show()
```
